```python
def tahun_kabisat(Tahun):
    if (tahun % 4 == 0 and tahun % 100 != 0) or (tahun % 400 == 0):
        return True
    else:
        return False

tahun = int(input("Masukkan tahun: "))
if tahun_kabisat(tahun):
    print(f"{tahun} adalah tahun kabisat.")
else:
    print(f"{tahun} bukan tahun kabisat.")
```

```python
angka1 = float(input("Masukkan angka pertama: "))
angka2 = float(input("Masukkan angka kedua: "))
angka3 = float(input("Masukkan angka ketiga: "))

print(f"Angka pertama: {angka1}")
print(f"Angka kedua: {angka2}")
print(f"Angka ketiga: {angka3}")

terbesar = max(angka1, angka2, angka3)

print(f"Angka yang paling besar diantara ketiganya adalah: {terbesar})
```

```python
def hitung_total_bayar(total_belanja, is_member):
    diskon = 0

    if total_belanja > 200_000:
        diskon = 4
    elif total_belanja > 100_000:
        diskon = 3

    if is_member:
        diskon += 2

    total_diskon = total_belanja * (diskon / 100)
    total_bayar = total_belanja - total_diskon

    return total_bayar

total_belanja = float(input("Masukkan total belanja: "))
is_member = input("Apakah konsumen adalah member? (y/n): ").lower() == 'y'

total_bayar = hitung_total_bayar(total_belanja, is_member)
print(f"Total yang harus dibayarkan: Rp {total_bayar}")
```

```python
def buat_user_baru():
    username = input("Masukkan username: ").strip()
    email = input("Masukkan email: ").strip()
    password = input("Masukkan password: ").strip()

    if not username:
        print("Username tidak boleh kosong!")
        return False

    if "@" not in email:
        print("Email tidak valid! Harus mengandung '@'.")
        return False

    if len(password) <= 8:
        print("Password harus lebih dari 8 karakter!")
        return False

    print("User baru berhasil dibuat!")
    print(f"Username: {username}")
    print(f"Email: {email}")
    return True

buat_user_baru()
```
