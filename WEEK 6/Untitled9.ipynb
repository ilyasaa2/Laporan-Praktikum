```python
tinggi = int(input("Masukkan jumlah baris: "))
for i in range (tinggi):
  print(' ' *(tinggi - i - 1), end='')
  print('* ' * (i + 1))
```

**Weekly Exercise 3**

```python
def pascal_row(n):
    row = [1]
    for k in range(1, n):
        row.append(row[k-1] * (n-k) // k)
    return row

n = int(input("Masukkan angka: "))
result = pascal_row(n)
print("Baris Pascal:", " ".join(map(str, result)))
total_sum = sum(result)
print(f"Jumlah total elemen pada baris Pascal: {total_sum}")
```
