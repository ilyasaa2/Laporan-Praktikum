```python
# main.py

# Import modul matematika
import matematika

# Menghitung luas lingkaran
jari_jari = float(input("masukkan jari jari lingkaran: "))
luas_lingkaran = matematika.luas_lingkaran(jari_jari)
print(f"Luas lingkaran dengan jari-jari {jari_jari} adalah {luas_lingkaran}")

# Menghitung luas persegi
sisi = float(input("masukkan luas persegi: "))
luas_persegi = matematika.luas_persegi(sisi)
print(f"Luas persegi dengan sisi {sisi} adalah {luas_persegi}")
```

```python
import math

# Menggunakan fungsi math.sqrt untuk menghitung akar kuadrat
angka = 25
akar = math.sqrt(angka)
print(f"Akar kuadrat dari {angka} adalah {akar}")

# Menggunakan fungsi math.pow untuk menghitung pangkat
basis = 2
eksponen = 3
pangkat = math.pow(basis, eksponen)
print(f"{basis} pangkat {eksponen} adalah {pangkat}")

# Menggunakan fungsi math.sin untuk menghitung sinus dari sudut (dalam radian)
sudut_radian = math.pi / 2
sinus = math.sin(sudut_radian)
print(f"Sinus dari 90 derajat adalah {sinus}")

# Menggunakan fungsi math.factorial untuk menghitung faktorial dari sebuah angka
angka = 5
faktorial = math.factorial(angka)
print(f"Faktorial dari {angka} adalah {faktorial}")

# Menggunakan fungsi math.ceil untuk membulatkan ke atas
angka = 4.3
ceil_angka = math.ceil(angka)
print(f"Pembulatan ke atas dari {angka} adalah {ceil_angka}")
```
