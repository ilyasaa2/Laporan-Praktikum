**Weekly Exercise 1**

```python
import math

def luas_lingkaran(radius):
    return math.pi * radius * radius

def luas_persegi(sisi):
    return sisi * sisi

def luas_segitiga(alas, tinggi):
    return 0.5 * alas * tinggi

def input_float(prompt):
    while True:
        try:
            value = input(prompt)
            if not value.strip():
                raise ValueError("tidak boleh kosong")
            return float(value)
        except ValueError as e:
            if str(e) == "tidak boleh kosong":
                print(e)
            else:
                print("hanya boleh memasukkan angka")

while True:
    print("\nKalkulator Luas Bangun Datar")
    print("1. Lingkaran")
    print("2. Persegi")
    print("3. Segitiga")
    print("4. Selesai")

    pilihan = input("Pilih bangun datar (1/2/3/4): ").strip()

    if pilihan == '1':
        try:
            radius = input_float("Masukkan jari-jari lingkaran: ")
            print(f"Luas lingkaran adalah: {luas_lingkaran(radius)}")
        except Exception as e:
            print(f"Terjadi kesalahan: {e}")
    elif pilihan == '2':
        try:
            sisi = input_float("Masukkan panjang sisi persegi: ")
            print(f"Luas persegi adalah: {luas_persegi(sisi)}")
        except Exception as e:
            print(f"Terjadi kesalahan: {e}")
    elif pilihan == '3':
        try:
            alas = input_float("Masukkan alas segitiga: ")
            tinggi = input_float("Masukkan tinggi segitiga: ")
            print(f"Luas segitiga adalah: {luas_segitiga(alas, tinggi)}")
        except Exception as e:
            print(f"Terjadi kesalahan: {e}")
    elif pilihan == '4':
        print("Program selesai. Terima kasih!")
        break
    else:
        print("Pilihan tidak valid. Silakan coba lagi.")
```

**Weekly Exercise 2**

```python
while True:
    try:
        NPM = input("Masukkan NPM: ").strip()

        if not NPM:
            raise ValueError("Input tidak boleh kosong.")
            continue

        if not NPM.isdigit():
            raise ValueError("Hanya boleh memasukkan angka.")
            continue

        jumlah_bilangan_genap = 0
        jumlah_bilangan_ganjil = 0

        for digit in NPM:
            angka = int(digit)
            if angka == 0:
                continue
            if angka % 2 == 0:
                jumlah_bilangan_genap += 1
            else:
                jumlah_bilangan_ganjil += 1

        print("Jumlah bilangan genap:", jumlah_bilangan_genap)
        print("Jumlah bilangan ganjil:", jumlah_bilangan_ganjil)
        break

    except ValueError as e:
        print(e)
        print("Silakan coba lagi.")
```
