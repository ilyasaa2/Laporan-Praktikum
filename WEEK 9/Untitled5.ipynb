**Weekly Exercise 1**

```python
1. List

Digunakan untuk menyimpan koleksi data yang dapat diubah.

Menggunakan tanda kurung siku [] untuk membuat list.

Elemen di dalam list bersifat terurut dan bisa memiliki duplikat.

Contoh: list_example = [1, 2, 3, 4]



2. Tuple

Digunakan untuk menyimpan koleksi data yang tidak dapat diubah.

Menggunakan tanda kurung () untuk membuat tuple.

Elemen di dalam tuple bersifat terurut dan bisa memiliki duplikat.

Contoh: tuple_example = (1, 2, 3, 4)



3. Set

Digunakan untuk menyimpan koleksi data yang unik dan tidak terurut.

Menggunakan tanda kurung kurawal {} untuk membuat set.

Elemen di dalam set tidak terurut dan tidak memiliki duplikat.

Contoh: set_example = {1, 2, 3, 4}



4. Dictionary

Digunakan untuk menyimpan pasangan kunci-nilai (key-value pair).

Menggunakan tanda kurung kurawal {} untuk membuat dictionary dengan format key: value.

Elemen di dalam dictionary bersifat tidak terurut, tetapi key harus unik.

Contoh: dict_example = {'key1': 'value1', 'key2': 'value2'}
```

**Weekly Exercise 2**

```python
# Data mahasiswa menggunakan tuple
data_mahasiswa = (
    "Nama Mahasiswa",  # Nama
    "Kelas Mahasiswa",  # Kelas
    "Tanggal Lahir Mahasiswa",  # Tanggal Lahir
    "No HP Mahasiswa",  # No HP
    "Alamat Mahasiswa",  # Alamat
    ("Hobi1", "Hobi2", "Hobi3")  # Hobi
)

# Menampilkan seluruh data dengan teknik perulangan
labels = ["Nama", "Kelas", "Tanggal Lahir", "No HP", "Alamat", "Hobi"]
for i, label in enumerate(labels):
    if label == "Hobi":
        print(f"{label}:")
        for hobi in data_mahasiswa[i]:
            print(f"  - {hobi}")
    else:
        print(f"{label}: {data_mahasiswa[i]}")
```
